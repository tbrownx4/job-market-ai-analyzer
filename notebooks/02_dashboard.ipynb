```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('jobs.csv')
df['position'] = df['position'].astype(str).str.strip()
df['company'] = df['company'].str.strip()
df['location'] = df['location'].str.strip()

df['Short_position'] = df['position'].apply(lambda x: ' '.join(x.split()[:5]))
df.head()
```

This chart shows the companies with the highest number of job listings in the dataset. It helps identify which organizations are hiring the most for data-related roles.

```python
top_companies = df['company'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_companies.plot(kind='bar')
plt.title('Top 10 Companies by Number of Job Listings')
plt.xlabel('Company')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Insight**
Amazon.com appears most frequently in the dataset, followed by companies like Microsoft and Google.
This suggests large tech companies dominate hiring in this job market snapshot.

## Top Cities for Job Listings

This chart highlights the cities with the highest concentration of job postings.
It gives insight into geographic hubs for data and technology roles.

```python
top_cities = df['location'].value_counts().head(10)
plt.figure(figsize=(10, 6))
top_cities.plot(kind='bar')
plt.title('Top 10 Cities by Number of Job Listings')
plt.xlabel('City')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Insight**
Cities such as Seattle, New York, and Boston show the highest number of listings, indicating strong regional demand for data-related jobs.

## Most common Job Titles

This chart displays the most frequently occuring job titles in the dataset.
Job titles were shortened for readability.

```python
top_short_positions = df['Short_position'].value_counts().head(10)

plt.figure(figsize=(10, 6))
top_short_positions.plot(kind='bar')
plt.title('Top 10 Job Titles (Shortened) by Number of Listings')
plt.xlabel('Job Title')
plt.ylabel('Number of Jobs')
plt.xticks(rotation=45, ha='right')
plt.show()
```

**Insight**
Roles like Data Scientist and Data Analyst appear more often, showing strong demand for analytical and data focused positions.
